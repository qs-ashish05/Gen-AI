```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
model_id ="google/gemma-3-1b-it"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id)  # loads the tokenizer of that specific model
```

```python
input_prompt = [
    "What is the capital of india"
]
```

```python
tokenized_inputs = tokenizer(input_prompt, return_tensors="pt")
print(tokenized_inputs)
print(tokenized_inputs["input_ids"])
```

```python

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype = torch.float16
)
```

```python
# fit the tokens into model
gen_result = model.generate(tokenized_inputs["input_ids"], max_new_tokens=25 )
print(gen_result)
```

```python
output = tokenizer.batch_decode(gen_result)
print(output)
```

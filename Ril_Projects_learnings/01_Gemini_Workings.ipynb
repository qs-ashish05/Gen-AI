## Gemini Intro
* Bard or gemini is a chatbot similar to chatGPT
* ChatGPT is based on the GOT3.5 and its varients
* Gemini is based on the google gemini pro & gemini pro vision
* Gemini is family of multimodel LLMs developed by the google deep minds
* Multimodel means it can perform task like -
    - Homogenous task (text to text generation or Image to image, etc)
    - Heterogenous task (text to image or image to text etc)

## Gemini History
1. Gemini is the successor of LaMDA and PaLM
2. LaMDA
    - Langauage model for Dialogue application
    - Released in 2021 to 2023
    - Model size range from 2B To 137B parameters
    - it was trained onver 1.56T words data (public dialouge data)

3. PaLM
    - pathway language Model
    - released in 2022

4. Gemini LLM
    - The central idea of the gemini is taken from the LaMDA and PaLM only.
    - gemini is an multii model

5. Three varient of the Gemini
    - gemini nano (for very basic uses like text generation from google search)
    - gemini pro ad gemini vision
    - gemini ULTRA

6. Features of Gemini pro
    - gemini pro is mostly used model and this was the actual model that was used in the gemini chatbot
    - most used LLM model of the google
    - gemini pro is equivalent to the GPT3.5 or GPT4 models

7. Features of Gemini ULTRA
    - based on the concept of AGI (Artificial Genral Inteligence)
    - Based on adaptive learning

```python
import google.generativeai as genai
import pathlib
import textwrap
from IPython.display import Markdown
from IPython.display import display
import os
```

```python
def to_markdown(text: str):
    text = text.replace("•", "*")
    return Markdown(textwrap.indent(text, '>', predicate=lambda line: True))
```

```python
text = "This is a sample text with •"
result = to_markdown(text)
display(result)   # do not use the print
```

```python
# Selecting a model to work with
model = genai.GenerativeModel('gemini-2.5-flash')
```

```python
%%time
response = model.generate_content("What is apple as fruit")
```

```python
res = to_markdown(response.text)
display(res)
```

```python
# the llm is generating the text response step by step i.e. streaming and we can access the each stream while it has been generating
# for that we use the following code
%%time
response = model.generate_content("What is cricket and football?", stream=True)
for chunk in response:
  print(chunk.text)
  print('__'*80)
```
